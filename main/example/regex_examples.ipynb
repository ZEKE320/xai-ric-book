```python
import nltk
from nltk import FreqDist
from nltk.corpus import gutenberg

nltk.download("all")
```

```python
text = gutenberg.raw("shakespeare-hamlet.txt")
print(text[:1000])
```

```python
words = nltk.word_tokenize(text)
print(words)
```

```python
pos = nltk.pos_tag(words)

print(pos)
```

```python
import re

sample_txt: str = """
This is an example sentence, but it is not the only one.
This is another example, and it is also valid.
This is a third example, so it should be included.
This is a fourth example, therefore it should be included as well.
"""

pattern: str = r"[^.!?]*\b(but|and|so)\b[^.!?]*[,.!?]"
matches = re.findall(pattern, sample_txt)
match_list = re.finditer(pattern, sample_txt)

for idx, match in enumerate(match_list):
    print(f"#{idx + 1}: {matches[idx]}")
    start = match.start()
    end = match.end()
    print(f"{sample_txt[start:end]}".replace("\n", " "))
```
